```python
#data e hora

def changing_region_time(hour, day):
  hour -= 3
  if hour < 0:
    hour += 24
    return str(hour), str(day-1)
  return str(hour), str(day)

name_of_day = ("Quinta", "Sexta", "Sábado", "Domingo", "Segunda", "Terça")

time_info = datetime.datetime.now()
hour, day = changing_region_time(time_info.hour, time_info.day)
```
